```
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```
books = pd.read_csv("Books.csv")
rating = pd.read_csv("Ratings.csv")
users = pd.read_csv("Users.csv")
```

```
books.head()
```

```
users.head()
```

```
rating.head()
```

```
books = books[["ISBN", "Book-Title", "Book-Author", "Year-Of-Publication", "Publisher"]]
books.rename(columns = {'Book-Title' : 'title', 'Book-Author' : 'author', 'Year-Of-Publication': 'year', 'Publisher' : 'publish'}, inplace=True)
```

```
books.head()
```

```
print("books shape ::", books.shape)
print("users shape ::", users.shape)
print("rating shape ::", rating.shape)
```

```
x =  rating['User-ID'].value_counts() > 200
user_value = x[x].index
user_value.shape
```

```
ratings = rating[rating["User-ID"].isin(user_value)]
```

```
ratings
```

```
books_rating = ratings.merge(books, on = "ISBN")
books_rating.rename(columns = {"Book-Rating" : "rating"}, inplace=True)
```

```
n_ratings = books_rating.groupby("title")["rating"].count().reset_index()
n_ratings.rename(columns = {"rating" : "n_ratings"}, inplace=True)
```

```
n_ratings.head()
```

```
final_data = books_rating.merge(n_ratings, on = "title")
```

```
final_data
```

```
final_ratings = final_data[final_data["n_ratings"] >=50]
```

```
final_ratings.drop_duplicates(['User-ID','title'], inplace=True)
```

```
final_ratings.shape
```

```
pivot_rating = final_ratings.pivot_table(columns = "User-ID", index = "title", values = "rating")
```

```
pivot_rating.fillna(0, inplace=True)
```

```
pivot_rating
```

```
from scipy.sparse import csr_matrix
sparse_rating = csr_matrix(pivot_rating)
```

```
print(sparse_rating)
```

```
from sklearn.neighbors import NearestNeighbors
```

```
NN = NearestNeighbors(algorithm = "brute")
```

```
NN.fit(sparse_rating)
```

```
Distance, suggestion = NN.kneighbors(pivot_rating.iloc[320,:].values.reshape(1,-1))
```

```
suggestion
```

```
for i in range(len(suggestion)):
  print(pivot_rating.index[suggestion[i]])
```

```
Distance, suggestion = NN.kneighbors(pivot_rating.iloc[237,:].values.reshape(1,-1))
```

```
for i in range(len(suggestion)):
  print(pivot_rating.index[suggestion[i]])
```

```
def suggestion_book(user_id):
  _, suggestion = NN.kneighbors(pivot_rating.iloc[user_id,:].values.reshape(1,-1))
  for i in range(len(suggestion)):
    print(pivot_rating.index[suggestion[i]])
```

```
suggestion_book(237)
```

```
suggestion_book(320)
```
